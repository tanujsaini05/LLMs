```python
from transformers import pipeline
from PIL import Image

# Load the document image using PIL
document_image = Image.open("C:/Users/tjwin/Downloads/ml.png")

# Define the question
question = "Who is the host of the webinar?"

# Load the document-question-answering pipeline
nlp = pipeline(
    "document-question-answering",
    model="impira/layoutlm-document-qa",
)

# Get the answer by passing the question and image
result = nlp(question=question, image=document_image)
print(result)
```
